```python
import numpy as np
import the_well
from the_well.data import WellDataset
from the_well.utils.download import well_download
base_path = "./data/WellData/"
```

```python
well_download(base_path=base_path, dataset="MHD_64", split="test", first_only=True)
```

```python
# --- 2. Load dataset ---
dataset = WellDataset(
    well_base_path='/Users/thomasmichel/Documents/EPFL/MA3/Project_DownScaling_CNN/space_time_pde_4d/experiments/rb2d/data/WellData',
    well_dataset_name="MHD_64",
    well_split_name="test",
    n_steps_input=1,
    n_steps_output=1,
    use_normalization=False,
)

ds = dataset.to_xarray()

# --- 3. Extract full arrays ---
# Velocity has shape: (time, x, y, z, 3)
vel = np.asarray(ds.velocity)     # float32 array
prs = np.asarray(ds.density)     # (sample, time, x, y, z)
vel = vel.reshape(-1, vel.shape[2], vel.shape[3], vel.shape[4], 3)
prs = prs.reshape(-1, prs.shape[2], prs.shape[3], prs.shape[4])

# Split velocity components
u = vel[..., 0]    # x-direction component
v = vel[..., 1]    # y-direction component
w = vel[..., 2]    # z-direction component

# Check shapes for sanity
print("u shape:", u.shape)
print("v shape:", v.shape)
print("w shape:", w.shape)
print("p shape:", prs.shape)

# --- 4. Save everything to a single .npz file ---
np.savez(
    "MHD_64_val.npz",
    u=u,
    v=v,
    w=w,
    p=prs
)

print("Saved MHD_64_val.npz successfully!")
```
